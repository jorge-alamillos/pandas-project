```python
#Importing the dataframe from attacks.csv

df = pd.read_csv('input/attacks.csv', encoding = 'latin-1') 
```

```python
#Displaying the columns in the dataframe
df.columns
```

```python
#Renaming some columns to express their content
df.rename(columns ={"Case Number": "Case_Number",
                    "Fatal (Y/N)": "Fatal_(Y/N)",
                    "Species ":"Species",
                    "Investigator or Source":"Investigator_or_Source",
                    "href formula": "PDF_Link",
                    "href": "PDF_Link_1",
                    "Case Number.1":"Case_Number",
                    "Case Number.2":"Case_Number_1",
                    "original order": "Original_Order"
                       } , inplace = True)

#Removing spaces from begining and end for every column
df.columns = df.columns.str.strip()

display(df.head())
```

Some columns seem to be duplicated so we are comparing the content:
- PDF_Link and PDF_Link_1,
- Case_Number and Case_Number_1 

```python
df_pdf = df[["PDF_Link","PDF_Link_1"]]
display(df_pdf.head())
```

```python
#Adding a column "compare" True if PDF_link = PDF_Link_1
df_pdf["compare"] = np.where(df["PDF_Link"] == df["PDF_Link_1"],"True", "False")

print(df_pdf["compare"].value_counts())
```

```python
#Filtering by compare = False(this means PDF_Link and PDF_Link_1 do not contain the same info).
#Also filtering to eliminate NULL values

df_pdf = df_pdf.loc[(df_pdf["compare"] == "False") & (~df_pdf["PDF_Link"].isnull()) & (~df_pdf["PDF_Link_1"].isnull())]
#Printing nº of items after filter
print(df_pdf["compare"].value_counts())
display(df_pdf)
#
```
